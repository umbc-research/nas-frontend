```python
from astropy.io import fits
fits_file = fits.open('20_59_40_HD_335264_00010.fits')
print(fits_file)
```

```python
image_data = fits_file[0].data
print(image_data)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
plt.figure()
plt.imshow(image_data, origin = 'lower', norm = LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```

```python
header = fits_file[0].header


instrument_name = header.get('INSTRUME', 'N/A')
filter_name = header.get('FILTER', 'N/A')  
date_name = header.get('DATE-AVG', 'N/A')
object_name = header.get('OBJECT', 'N/A')

print("Object Name:", object_name)
print("Instrument Name:", instrument_name)
print("Filter:", filter_name)
print("Date-avg:", date_name)
```
